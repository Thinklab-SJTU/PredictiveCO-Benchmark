```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def plot_curve(df, symbol):
    curve = df[df.Symbol == symbol]
    sns.lineplot(x =np.arange(len(curve)), y=curve['Close'])
    plt.show()
```

```python
model_names = ["mse", "dfl", "blackbox", "identity", "qptl", "spo", "nce", "pointLTR", "pairLTR", "listLTR", "lodl"]
prob_name = "knapsack-gen"
prefix = "default"
results_list = []
for model_name in model_names:
    try:
        res = np.load(f"/mnt/nas/home/genghaoyu/OR/PTO/Rethink1.0/saved_records/{prob_name}/{model_name}/{prefix}/results.npy")
        size = res.shape
    except:
        res = np.ones(size)
        res[1] = 0.1
    print(res.shape)
    results_list.append(res)
```

```python
print(len(results_list), results_list[0][1]/results_list[0][0])
```

```python
results_list = np.array(results_list)
```

```python
regret_list = np.array([res[1]/res[0] for res in results_list])
print(regret_list.shape)
```

```python
regret_dict = {}
formal_names = ["TwoStage", "DFL", "Blackbox", "Identity", "QPTL", "SPO", "NCE", "Pointwise LTR", "Pairwise LTR", "Listwise LTR", "LODL"]
for idx in range(len(model_names)):
    regret_dict[formal_names[idx]] = regret_list[idx]
```

```python
regret_dict.keys()
```

```python
import pandas as pd
df = pd.DataFrame(regret_dict)
```

```python
# 创建一个箱线图
# plt.figure(figsize=(10, 6))
# ax = sns.boxplot(x=df['columns'], y=df.iloc[1], palette="Set2")
plt.figure()
ax = sns.boxplot(data = df, orient="h", palette="Set2")

# 可选：旋转x轴标签
# ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.title("Boxplot with Standard Deviation")
plt.show()
```
